- Fonction **randint()**


```randint(min,max)``` permet de générer un nombre entier pseudo-aléatoire entre min et max.

````python
from random import randint
a = randint(10,20)
print(a)
````

### Correction de l'étape 1

### Étape 2

Toujours en demandant **une** multiplication aléatoire, vérifier la réponse donnée par l'utilisateur et reposer la même question tant que la bonne réponse n'est pas donnée.

On pourra utiliser une boucle infinie ```while True``` dont on s'éjectera par ```break```.

````python
from random import randint
a = randint(1,10)
b = randint(1,10)
question = str(a) + "*" + str(b) + "="
while True :
  reponse = int(input(question))
  if reponse == a*b :
    print('bravo')
    break
  else :
    print('erreur')
````

### Étape 3

- Poser successivement 5 opérations.
- Poser successivement n opérations, où n est un nombre rentré par l'utilisateur.

````python
from random import randint

n = int(input("Combien d'opérations voulez-vous ?"))
for k in range(n):
  a = randint(1,10)
  b = randint(1,10)
  question = str(a) + "*" + str(b) + "="
  while True :
    reponse = int(input(question))
    if reponse == a*b :
      print('bravo')
      break
    else :
      print('erreur')
````

````python
from random import randint
import time

n = int(input("Combien d'opérations voulez-vous ?"))
print("c'est parti !")
chrono_start = time.time()
for k in range(n):
  a = randint(1,10)
  b = randint(1,10)
  question = str(a) + "*" + str(b) + "="
  while True :
    reponse = int(input(question))
    if reponse == a*b :
      print('bravo')
      break
    else :
      print('erreur')
temps_total = time.time()-chrono_start
print("Terminé. Vous avez mis {:.3} secondes".format(temps_total))
````
